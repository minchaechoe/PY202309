```python
def add_task(task_list, task):
    task_list.append(task) # add a new task in task_list by append function

def display_tasks(task_list):
    if not task_list:
        print("할 일 목록이 비어 있습니다.")
    else:
        print("할 일 목록:")
        for i in range(len(task_list)): # print task_list by repeatition
            task = task_list[i]
            print(f"{i+1}. {task}")
    

def save_tasks_to_file(task_list, filename):
    write_fp = open(filename, 'w', encoding="utf8")
    for task in task_list:
        write_fp.write(task+'\n') # write tasks in task_list to file
    write_fp.close()

def load_tasks_from_file(filename):
    task_list = []
    read_fp = open(filename, 'r', encoding='utf8')
    lines = read_fp.readlines() # read the all lines in flie, and save in list
    for line in lines:# print the file line by line
        task_list.append(line.strip()) # strip : removes leading and trailing whitespace
    return task_list

file_name = "tasks.txt"
task_list = load_tasks_from_file(file_name)

while True:
    print("\n할일 관리 프로그램")
    print("1. 할일 추가")
    print("2. 할일 목록 표시")
    print("3. 종료")
    
    choice = input("선택: ")
    
    if choice == '1':
        task = input("할일을 입력하세요: ") # input task by user
        add_task(task_list, task)
        print("할일이 추가되었습니다.")
    elif choice == '2':
        display_tasks(task_list)
    elif choice == '3':
        save_tasks_to_file(task_list, file_name)
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 선택입니다. 다시 시도하세요.")
```
